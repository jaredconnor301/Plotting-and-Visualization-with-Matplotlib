# Colors, Markers, and Line Styles

#### Colors:
We can also alter the X & Y coordinates' colors and line types with various strings. For example, to plot x & y with green dashes we could pass `ax.plot(x, y, 'g--')`. If we prefer not to short cut the specific abbreviations, we could explicitly state `ax.plot(x, y, linestyle='--', color='g')`. Same plot, just more explicitly stating the style.

#### Markers:
Line plots can additionally have _markers_ to highlight their actual data points. The market can be part of the style string that highlights a given point on the line. 

```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np
from numpy.random import randn
```

```python
data = randn(30).cumsum()
```

```python
ax1 = plt.figure()
ax1 = plt.plot(data)
```

This could more explicitly have been written as:

```python
ax2 = plt.figure()
ax2 = plt.plot(data, color='g', linestyle='dashed', marker='o')
```

```python
ax3 = plt.figure()
ax3 = plt.plot(data, 'p')
```
